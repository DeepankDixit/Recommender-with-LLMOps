- Create .env file

    - GROQ_API_KEY=""

    - HUGGINGFACEHUB_API_TOKEN=""

- Create the project structure
    - data\ - to contain the data

- Create requirements.txt file

- Create setup.py to install all requirements, dependencies, and packages (https://github.com/data-guru0/ANIME-RECOMMENDER-SYSTEM-LLMOPS/blob/main/setup.py)
    - How to trigger/run your setup.py :
        
        pip install -e .

- Create src\ folder
    - Since src\ will be treated as a package, must create an __init__.py inside it
    - src\ will be used to store main components - data loader, prompt templates, recommender, and vector store etc

- Create utils\ folder for utilities 
    - for helper functions, logger functions, custom exception function
    - Since we'd also wanna import functions from utils\ folder, we have to create this too as a package -> hence, create __init__.py file inside this too
    - copy the code for both folders from the github repo

-  Create config\ folder for configurations 
    - contains API setup, hugging face setup
    - database setup
    - config\ will also be treated as a package, so gotta create __init__.
    
- Create app\ folder for containing our streamlit application

- Create pipeline\ folder for keeping two pipelines 
    - training pipeline
    - recommendation / inference / prediction pipeline
    - while in src\ we keep all components of creating the application, in pipeline\ we join all components to form a pipeline
    - pipeline\ will also be treated as a package, so gotta create __init__.

- We now again run the "pip install -e ." command 
    - because we have created packages -
        - the data\ directory is a folder,
        - but the src\, utils\, config\, app\, pipeline\ directories are packages
